```python

# 1. INSTALL NECESSARY LIBRARIES
!pip install streamlit scikit-learn pandas matplotlib seaborn -q
!npm install -g localtunnel -q

import pandas as pd
import streamlit as st
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# 2. CREATE THE STREAMLIT APP FILE
# This writes the dashboard code into a file named 'app.py'
with open('app.py', 'w') as f:
    f.write("""
import streamlit as st
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Page Configuration
st.set_page_config(page_title="Maternal Health Dashboard", layout="wide")

# Load Data
@st.cache_data
def load_data():
    file_id = '1-H5uC2M9RgE6Q92_3LxOa5_q-8z4PeDF'
    url = f'https://drive.google.com/uc?export=download&id={file_id}'
    df = pd.read_csv(url)
    risk_mapping = {'low risk': 0, 'mid risk': 1, 'high risk': 2}
    df['RiskLevel'] = df['RiskLevel'].map(risk_mapping)
    return df

df = load_data()

# Train Model
X = df.drop('RiskLevel', axis=1)
y = df['RiskLevel']
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# Dashboard Layout
st.title("ūü§įūüŹĽ Maternal Health Risk Prediction")
st.markdown("Enter patient vitals in the sidebar to get an instant risk assessment.")

# Sidebar Inputs
st.sidebar.header("Patient Vital Signs")
age = st.sidebar.slider("Age", 10, 60, 25)
sys_bp = st.sidebar.slider("Systolic BP", 70, 160, 120)
dia_bp = st.sidebar.slider("Diastolic BP", 50, 100, 80)
bs = st.sidebar.slider("Blood Sugar (mmol/L)", 6.0, 20.0, 7.5)
temp = st.sidebar.slider("Body Temp (¬įC)", 35.0, 40.0, 37.0)
hr = st.sidebar.slider("Heart Rate", 50, 120, 70)

# Prediction Logic
input_data = pd.DataFrame([[age, sys_bp, dia_bp, bs, temp, hr]],
                          columns=['Age', 'SystolicBP', 'DiastolicBP', 'BS', 'BodyTemp', 'HeartRate'])
prediction = model.predict(input_data)
risk_labels = {0: "Low Risk", 1: "Mid Risk", 2: "High Risk"}
colors = {0: "#28a745", 1: "#ffc107", 2: "#dc3545"}

# Display Results
st.subheader("Prediction Analysis")
st.markdown(f"<h1 style='text-align: center; color: {colors[prediction[0]]};'>{risk_labels[prediction[0]]}</h1>", unsafe_allow_html=True)

# Add Feature Importance Chart
st.subheader("Key Influencing Factors")
importances = pd.Series(model.feature_importances_, index=X.columns).sort_values()
st.bar_chart(importances)
    """)

# 3. GET YOUR PUBLIC IP (Required for LocalTunnel)
import urllib
print("Password/Endpoint IP for the link below is:", urllib.request.urlopen('https://ipv4.icanhazip.com').read().decode('utf8').strip())

# 4. RUN THE SERVER (With port cleaning)
# Force kill any existing process on port 8501
!fuser -k 8501/tcp

# Run the Streamlit app
!streamlit run app.py & npx localtunnel --port 8501
```
